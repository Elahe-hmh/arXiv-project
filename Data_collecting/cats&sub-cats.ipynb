```python
import pandas as pd
df_long=pd.read_json('merged_14.json')
```

```python
def drop(df_long):
    dff=df_long.drop(['identifier','datestamp','setSpec','created','id'],axis=1)
    df=dff.drop(['updated','authors','comments','journal-ref'],axis=1)
    return df
```

```python
df=drop(df_long)
```

```python
def Cat_uni(df):
    cat=df['categories']
    cat_uni=[]
    for i in cat:
        if i not in cat_uni:
            cat_uni.append(i)
    return cat_uni        
```

```python
cat_uni=Cat_uni(df)
#cat_uni
```

```python
def space_split(cat_uni):
    cat_split_space=[]
    for i in cat_uni:
        if type(i)==str:
            e=i.split()
            cat_split_space.append(e)
    return cat_split_space        
    
```

```python
cat_split_space=space_split(cat_uni)
#cat_split_space
```

```python
def main(cat_split_space):
    main_cats=[]
    q=[]
    for k in cat_split_space:
        for t in k:
            q=t.split(".")
            main_cats.append(q[0])
    main_uni=[]
    for m in main_cats:
        if m not in main_uni:
            main_uni.append(m)
    return main_uni
    
```

```python
main_uni=main(cat_split_space)
#main_uni
```

```python
def sub_cat(cat_split_space):
    b=[]
    for k in cat_split_space:
        for t in k:
            q=t.split(".")
            if len(q)>1:
                b.append(q)
    return b     
```

```python
cat_sub_cat=sub_cat(cat_split_space)
cat_sub_cat
```

```python
def subs(cat_sub_cat):
    w=[]
    a=dict()
    for i in cat_sub_cat:
        if i[0] in a:
            s = a[i[0]]
            if i[1] not in s:
                s.append(i[1])
            a[i[0]] = s
        else:
            a[i[0]] = [i[1]]
    return a        
```
